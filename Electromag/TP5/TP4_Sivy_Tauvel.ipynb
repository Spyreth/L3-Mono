## I - Diffraction par des fentes
Dans cette première partie, nous nous intéresserons à la diffraction d'une lumière à vapeur de mercure par des fentes. 
La diffraction est un phénomène qui apparaît lorsque l'image d'un objet est perturbée par un ou plusieurs obstacles sur le chemin de l'onde. Cette déformation de l'image attendu n'est visible que lorsque la taille de l'objet déformant est de l'ordre de 100 fois la longueur d'onde.

Afin d'observer la diffraction, nous nous plaçons dans les conditions dites de Fraunhofer. Cette configuration consiste à créer une image à l'infini avec une lentille convergente sur un écran, d'un faisceau lumineux ponctuel diffracté. On utilisera donc :
- Un diaphragme en fente
- Une lampe à vapeur de mercure
- Une lentille convergente $L_0$ de focal $f'_0$
- Une lentille convergente $ L_1$ de focal $f'_1$

Pour obtenir cet image à l'infini, on placera notre diaphragme en fente devant la lampe et dans le plan focal de $L_0$. On placera ensuite uner autre lentille $L_1$ ainsi que l'écran dans le plan focal de $L_1$. 
Nous sommes désormais en configuration de Fraunhofer.

Ces conditions étant établis, nous pouvons dorénavant placer une fente calibrée entre les deux lentilles. Voir figure n°1. 
L'image observée sur l'écran n'a alors plus rien à voir. Tandis que la fente est verticale, on observe une série horizontale d'alternance de lignes verticales brillantes puis sombres. La luminosité est maximale sur la frange centrale et diminue au fur et à mesure qu'on s'écarte du milieu. 
On observe le même phénomène avec un cheveux qui remplace la fente calibrée. 

Cette image est étonnante car elle ne peut être expliquée avec les lois géométrique de la lumière particulaire. Nous devons considérer le faisceaux comme une onde que le passage par la fente va diffracter. On aura alors des zones d'amplitudes maximale où on observe l'image attendu de la fente, et des zones d'amplitudes minimale où on verra simplement du noir. 

On cherche, par la mesure de la frange brillante principale à determiner la taille de la largeur de la fente. On connait la fonction donnant l'intensité lumineuse de l'image dans le cas d'une fente unique. On se place à 1 dimension et on trouve : $I(x) = I_0sinc^2(\pi a f_X)$. Cette valeur s'annule pour un $\Delta x$ correspondant à la  distance verticale de la frange brillante principale. Le sinc quant à lui s'annule pour une valeur dans la parenthèse de $\pi$. On détermine ainsi notre équation : 
$$\frac{a\Delta x}{\lambda f'} = $$
$$\rightarrow a = \frac{\lambda f'}{\Delta x}$$

Comme on a placé un filtre gélatineux jaune, on sait que seul les longeur d'onde jaune peuvent passer. On choisira donc $\lambda = 578.51nm$, correspondant à la moyenne des deux raies jaunes de la lampe à vapeur de mercure.

Nous déterminons la focale de notre lentille à $f' = 30\pm 0.5 cm$. Cette valeur n'est pas donnée au hasard, il s'agit de notre estimations d'erreur lorsque nous avons utiliser la méthode d'autocollimation pour placer l'écran dans le plan focal de la lentille.

Enfin, on mesure $\Delta x = 0.962 \pm{3.10^{-3}} mm $

N'oublions pas du calcul la propagation des erreurs qui sera faites avec la méthode des moindres carrées


Ainsi, $a = 180 \pm{30} \mu m  $.


Nous plaçons maintenant comme objet diffractant un réseau de 2 fentes. 
On remarque que dans le cas de 2 fentes, on n'obtient pas excatement comme image la somme de la diffraction de chaque fente. En effet, en plus du phénomène de diffraction, les faisceaux passant par chacunes des fentes vont interférer les uns avec les autres. On obtient alors sur la figure des maximum d'amplitudes là où on n'en voyait pas pour la simple fentes. On cherche par la mesure de la figure obtenue à determiner la distance entre les deux franges. On sait que dans la formule donnant l'intensité lumineuse, la partie dépendant de la distance inter-frange 'b' est : $cos(2\pi b f_X)$. 
Avec $f_X = \frac{x}{\lambda f'}$ la fréquence spatiale

On mesure donc la distance inter-frange $\Delta x = 0.100\pm{67.10^{-4}mm}$ dont on déduit b le distance entre les fentes $b=\frac{\lambda f'}{4\Delta x}$

d'où $b= 0.43\pm 3.10^{-2} mm $

## II - Diffraction par un réseau 

Nous remplaçons maintenant la fente par un réseau. Nous retirons aussi la loupe de Fresnel pour y mettre un simple écran, que nous utiliserons comme spectromètre, C'est à dire que nous mesurerons la distance entre les franges de chaques couleurs, afin de déterminer la longueur d'onde en jeu.  

Dans un premier temps, nous chercherons à vérifier la loi donnant les angles pour lesquels le réseau diffracte qui est : $asin(\theta_n) = n\lambda$
où : 
- a est le pas du réseau
- $\lambda$ est la longueur d'onde
- $\theta$ est l'angle d'incidence
- n est un entier quelconque 

Vérifions la proportionnalité entre $sin \theta_n$ et n : 

```python
#importation des librairies 
import numpy as np
import pandas as pd #in case the porgram that get the data export's it in csv or an other python can't take by itself
import qexpy as q
import qexpy.plotting as qplt

l = np.array([0.3])
l_error = 0.0005
n = np.array([-2, -1, 0,1,2])
R = np.array([-11.6, -5.8, 0, 5.8, 11.5])*10**(-2)
error_R = 0.0005

R_q = q.MeasurementArray(R, error = error_R)
L_q = q.MeasurementArray(l, error = l_error)

sin_theta = R_q/(R_q**2 + L_q**2)**(0.5)


fig1 = qplt.plot(n, sin_theta, residuals=True) 

result = fig1.fit(model=q.FitModel.LINEAR, parguess = [2, 0])

figure = qplt.get_plot()
figure.title = 'Graphe du sin de l\'angle d\'incidence selon un entier n quelconque'

# Affichage du graphe et du fit
figure.show()
print(result)
```

On observe dans un premier temps qu'il existe bien une relation de proportionnalité entre l'entier relatif n et le sinus de l'angle d'incidence $sin\theta_n$. On remarque aussi que les erreurs sont à peine visible sur le graphe. En effet, la mise en place permet une lecture précise de l'angle, et l'entier n ne représente aucune source d'erreur possible, la propagation est donc faible.

Ce fit linéaire simple nous permet aussi de déterminer la longueur d'onde du faisceau ou encore le pas du réseau selon ce que l'on cherche car on retrouve une fonction affine de la forme : 
$sin(\theta_n)(n) = \frac{\lambda}{a}n$.

On peut ainsi verifier la proportionnalité entre $sin\theta_n$ et $\lambda$ en fixant l'entier n à 1.

On observe bien une comportement linéaire tel que : $sin(\theta_n) = \frac{\lambda}{a}n + b$

avec $\frac{\lambda}{a} = 0.181 \pm 0.002$ et $b = -0.001\pm0.003$


On peut discuter la précision de a et b, mais on remarque que les valeurs attendues sont comprise dans l'intervalle d'erreur. En effet, b devrait être nul car dans le cas où $n=0$, l'image est observée est direct, $\theta_0 = 0$ et on retrouve bien $b=0$.

La valeur $a$ quant à elle peut être déterminer par la valeur constructeur, dont on trouve $\frac{\lambda}{a} = 0.183$. 

On utilisera maintenant un spéctromètre à réseau. Cet outil permet de disperser spatialementnles longueurs d'onde sur un capteur. Il faut cependant calibrer le spectromètre afin de connaître la relation définissant la longueur d'onde slon la position affciher sur le capteur. Connaissant les raies de la lampe à vapeur de mercure, nous pouvons ainsi l'étalonner.

```python
lambd = np.array([579.07, 576.96, 435.83, 404.66, 365.02])
x = np.array([58, 58,43, 40, 36])

x_error = 5

x_q = q.MeasurementArray(x, error = x_error, name = 'distance verticale au centre en mm')
fig1 = qplt.plot(x, lambd, residuals=True) 
result = fig1.fit(model=q.FitModel.LINEAR, parguess = [2, 0])
figure = qplt.get_plot()
figure.error_bars(True)
figure.title = 'Graphe de la longueur d\'onde en fonction de la position x sur le spectromètre'

# Affichage du graphe et du fit
figure.show()
print(result)
```

On a donc défini un origine au centre tel que $x=0$ sur ce point, puis déterminer la distance de chaque raies avec cet origine. ce tableau établi, on peut tracer le courbe affine reliant ces point et ainsi, établir la courbe d'étalonnage de notre spectromètre-fictif. 

On observe bien une courbe affine comme on s'y attendait, donnant la longueur d'onde selon la position. On retrouve ainsi $\lambda(x) = Ax + B$ avec 

$A = 9.64 \pm 0.08$ et $B = 19\pm4$

Les grandeurs A et B sont sans dimension.

Ceci nous permet ainsi de déduire la longueur d'onde de la raies verte selon sa position avec la fonction suivante : $\lambda_{vert} = (9.64\pm0.08)(54\pm0.5) + (19\pm4)$

Nous utiliserons la méthode des moindres carrés pour la propagation des erreurs.

D'où $\lambda_{vert} = 539 \pm 4,99 nm$

Cette valeur est tout à fait cohérentes car elle se situe bien dans le spectre visible au niveau du vert.
